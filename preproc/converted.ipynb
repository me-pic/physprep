```python
from phys2bids.phys2bids import phys2bids
import argparse
import sys
import json
import pandas as pd
import gc
import os
import bioread
import matplotlib.pyplot as plt
```

```python
root = '/scratch/mepicard/emotionvideos/sourcedata/physio'
save = '/scratch/mepicard/physio_data/emotionvideos'
sub = ['sub-02', 'sub-03', 'sub-05']
```

```python
def load_json(file):
    f = open(file)
    data = json.load(f)
    f.close()
    return data
```

```python
for s in sub:
    print('-'*20 + s + '-'*20)
    sessions = [f for f in os.listdir(os.path.join(save, s)) if os.path.isdir(os.path.join(save,s,f))]
    for ses in sessions:
        print('-'*20 + ses + '-'*20)
        files = [f.split('.')[0] for f in os.listdir(os.path.join(save, s, ses))]
        files = [*set(files)]
        for file in files:
            print(file)
            try:
                header = load_json(os.path.join(save, s, ses , f'{file}.json'))['Columns']
                run = pd.read_csv(os.path.join(save, s, ses, f'{file}.tsv.gz'), sep='\t', names=header)
                plt.plot(run['TTL'])
                plt.show()
            except:
                continue
```
